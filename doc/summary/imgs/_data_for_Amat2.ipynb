```python
from exputils.vis_matrix import vis_matrix
from exputils.dot.recovery_state import recovery_states_from_idxs
from exputils.stabilizer_group import total_stabilizer_group_size
from exputils.state.random_ket import make_random_quantum_state
from exputils.extent.custom import calculate_extent_custom
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.sparse

n = 2

Amat = recovery_states_from_idxs(n, list(range(total_stabilizer_group_size(n))))
b = make_random_quantum_state("pure", n, seed=4)

se, x, y = calculate_extent_custom(n, Amat, b)
print(se)
A_sorted = Amat[:, np.argsort(np.abs(Amat.T @ b.conj()))[::-1]]
A_sorted_cut = A_sorted[:, :8]

se2, x2, y2 = calculate_extent_custom(n, A_sorted_cut, b)
print(se2)
A_sorted2 = Amat[:, np.argsort(np.abs(Amat.T @ y2.conj()))[::-1]]
A_sorted2_cut = A_sorted2[:, :1]
final_A = scipy.sparse.hstack([A_sorted_cut, A_sorted2_cut])

se3, x3, y3 = calculate_extent_custom(n, final_A, b)
print(se3)

vis_matrix(A_sorted_cut.toarray())
vis_matrix(final_A.toarray())
```

```python
cmap = matplotlib.colormaps["viridis"]
for idx, val in enumerate([1, 1 / np.sqrt(2), 1 / 2, -1 / 2, -1 / np.sqrt(2)]):
    rgb = cmap(int((val + 1) * 255 / 2))[:3]
    # print(idx, rgb)
    print(
        "\\definecolor{c"
        + str(idx)
        + "}"
        + f"{{RGB}}{{{int(rgb[0] * 255)}, {int(rgb[1] * 255)}, {int(rgb[2] * 255)}}}"
    )


for A in [Amat, A_sorted, A_sorted_cut, final_A]:
    B = A.toarray().real
    B[B == 0] = np.nan
    plt.imshow(B)
    plt.show()


    A = A.tocoo()

    val_to_idx = {}
    val_to_idx[1] = 0
    val_to_idx[1 / np.sqrt(2)] = 1
    val_to_idx[1 / 2] = 2
    val_to_idx[-1 / 2] = 3
    val_to_idx[-1 / np.sqrt(2)] = 4

    print("\\foreach \\color/\\col/\\row in {")
    for i, (val, row, col) in enumerate(zip(A.data, A.row, A.col)):
        if val.real != 0:
            assert val.real in val_to_idx
            idx = val_to_idx[val.real]
            print(f"c{idx}", col, -row, sep="/", end="")
        else:
            assert val.imag in val_to_idx
            idx = val_to_idx[val.imag]
            print(f"c{idx}", col, -row - (2**n) - 1, sep="/", end="")
        if i != A.nnz - 1:
            print(",")
        else:
            print("\n}{")
```
