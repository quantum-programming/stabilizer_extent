```python
import numpy as np

cQs = []
for _c in range(4):
    for _Q in range(8):
        c = [0] * 2
        Q = [[0, 0], [0, 0]]
        c[0] = (_c >> 1) & 1
        c[1] = _c & 1
        Q[0][0] = (_Q >> 2) & 1
        Q[0][1] = (_Q >> 1) & 1
        Q[1][1] = _Q & 1
        cQs.append((c, Q))

cQs.sort(key=lambda cQ: cQ[0][0] * 4 + cQ[1][0][0] * 2 + cQ[1][0][1])

idxs = [0, 1, 2, 3, 4, 5, 6, 7, 28, 29, 30, 31]

cnt = 0
diff = 0
ret = ""
for idx, (c, Q) in enumerate(cQs):
    print(f"idx={idx:<2}, c={c}, Q={Q}", end=" | ")

    if idx in idxs:
        ret += str(idx) + "/" + str(cnt + diff) + "/"
        cnt = (cnt + 1) % 4

    for _x in range(4):
        x = [0] * 2
        x[0] = (_x >> 1) & 1
        x[1] = _x & 1
        val = ((-1) ** (np.array(x).T @ np.array(Q) @ np.array(x))) * (
            (1j) ** (np.array(c).T @ np.array(x))
        )
        term = ""
        if val == 1:
            term += " +/"
        elif val == -1:
            term += " -/"
        elif val == 1j:
            term += "+i/"
        elif val == -1j:
            term += "-i/"
        print(term, end=" ")
        if idx in idxs:
            ret += term
    if idx in idxs:
        ret += ",\n" if idx != 31 else "\n"
        if idx == 3:
            diff = 5
        if idx == 7:
            diff = 16
    print()

print("-" * 10)
print(ret)
```
