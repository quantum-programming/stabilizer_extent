```python
import numpy as np
from exputils.Amat.get import get_Amat
from exputils.vis_matrix import vis_matrix
from exputils.dot.recovery_state import recovery_states_from_idxs
from exputils.stabilizer_group import total_stabilizer_group_size
import matplotlib.pyplot as plt
import matplotlib

n = 2
A = get_Amat(n)
vis_matrix(A)

A = recovery_states_from_idxs(n, list(range(total_stabilizer_group_size(n))))
vis_matrix(A.toarray())
```

```python
B = A.toarray().real
B[B == 0] = np.nan
plt.imshow(B)
plt.show()

cmap = matplotlib.colormaps["viridis"]
for idx, val in enumerate([1, 1 / np.sqrt(2), 1 / 2, -1 / 2, -1 / np.sqrt(2)]):
    rgb = cmap(int((val + 1) * 255 / 2))[:3]
    # print(idx, rgb)
    print(
        "\\definecolor{c"
        + str(idx)
        + "}"
        + f"{{RGB}}{{{int(rgb[0] * 255)}, {int(rgb[1] * 255)}, {int(rgb[2] * 255)}}}"
    )
```

```python
Is = []
for i, col in enumerate(A.T):
    if np.all(col.data.imag == 0):
        Is.append(i)
print("\\foreach \\i in {")
print(*Is, sep=",")
print("} {")
```

```python
A = A.tocoo()

val_to_idx = {}
val_to_idx[1] = 0
val_to_idx[1 / np.sqrt(2)] = 1
val_to_idx[1 / 2] = 2
val_to_idx[-1 / 2] = 3
val_to_idx[-1 / np.sqrt(2)] = 4

print("\\foreach \\color/\\col/\\row in {")
for i, (val, row, col) in enumerate(zip(A.data, A.row, A.col)):
    if val.real != 0:
        assert val.real in val_to_idx
        idx = val_to_idx[val.real]
        print(f"c{idx}", col, -row, sep="/", end="")
    else:
        assert val.imag in val_to_idx
        idx = val_to_idx[val.imag]
        print(f"c{idx}", col, -row - (2**n) - 1, sep="/", end="")
    if i != A.nnz - 1:
        print(",")
    else:
        print("\n}{")
```
